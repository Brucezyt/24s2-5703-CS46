```python
import subprocess
import yaml
import os
```

```python

yaml_file_path = r"C:\Users\Bruce\Desktop\5703 recouses\dataset\datayolo8-scaledown.yaml"

with open(yaml_file_path, 'r', encoding='gbk') as file:
    data_config = yaml.safe_load(file)


train_images = data_config.get('train')
val_images = data_config.get('val')
test_images = data_config.get('test')

train_labels = data_config.get('train_labels')
val_labels = data_config.get('val_labels')
test_labels = data_config.get('test_labels')

print("train_images_path:", train_images)
print("valid_images_path:", val_images)
print("test_images_path:", test_images)
print("train_labels_path:", train_labels)
print("valid_labels_path:", val_labels)
print("test_labels_path:", test_labels)
```

```python
def check_data_integrity(image_folder, label_folder):
    missing_labels = []
    image_files = os.listdir(image_folder)
    for image_file in image_files:
        image_name = os.path.splitext(image_file)[0]  
        label_file = f"{image_name}.txt"
        if not os.path.exists(os.path.join(label_folder, label_file)):
            missing_labels.append(image_file)
    return missing_labels


missing_train_labels = check_data_integrity(train_images, train_labels)
if missing_train_labels:
    print(f"Training set missing label file: {missing_train_labels}")
else:
    print("The training set label file is complete.")


missing_val_labels = check_data_integrity(val_images, val_labels)
if missing_val_labels:
    print(f"Valid set missing label file: {missing_val_labels}")
else:
    print("The valid set label file is complete.")
```

```python
train_command = [
    "python", "train.py",
    "--img", "640",  
    "--batch", "16",  
    "--epochs", "10",  
    "--data",r"C:\Users\Bruce\Desktop\5703 recouses\dataset\datayolo8-scaledown.yaml",  
    "--weights", "yolov8m.pt", 
    "--device", "0"  
]

subprocess.run(train_command, cwd=r'C:\Users\Bruce\Desktop\5703 recouses\yolov8')
```
